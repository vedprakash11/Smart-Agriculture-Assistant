```python
X_train, X_test, y_train, y_test = train_test_split(data, output, test_size=0.33, random_state=42)
from sklearn import linear_model
reg = linear_model.LinearRegression(fit_intercept=False,normalize=True,n_jobs=5)
```

```python
reg.fit(X_train,y_train)
```

```python
pred = reg.predict(X_test)
```

```python
from sklearn import metrics
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, pred)
```

```python
regr_1 = DecisionTreeRegressor(max_depth=6,random_state=8)
regr_1.fit(X_train, y_train)
```

```python
pred = regr_1.predict(X_test)
```

```python
r2_score(y_test, pred)
```
